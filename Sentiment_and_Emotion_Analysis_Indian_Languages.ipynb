### Assinging Emotions to Sentiment column using Zro-shot-classification model

```python
import pandas as pd
from transformers import pipeline
import torch
from tqdm import tqdm
import re

# Define emotion labels
emotion_labels = ["joy", "sadness","love", "anger", "fear", "surprise"]

# Define hypothesis templates for Hindi and Telugu
hypothesis_templates = {
    'hi': "यह वाक्य {} को दर्शाता है।",
    'te': "ఈ వాక్యం {} ను సూచిస్తుంది.",
}

# Load zero-shot classification model
emotion_classifier = pipeline(
    "zero-shot-classification",
    model="joeddav/xlm-roberta-large-xnli",
    tokenizer="joeddav/xlm-roberta-large-xnli",
    device=0 if torch.cuda.is_available() else -1
)

# Load your dataset
df = pd.read_csv("combined_df.csv")
sentences = df["Cleaned_Sentence"].fillna("").astype(str).tolist()

# Helper function to detect language based on Unicode range
def detect_language(text):
    hindi_pattern = r'[\u0900-\u097F]'
    telugu_pattern = r'[\u0C00-\u0C7F]'

    if re.search(telugu_pattern, text):
        return 'te'
    elif re.search(hindi_pattern, text):
        return 'hi'
    else:
        return 'hi'  # Default to Hindi if uncertain

# Output list for predicted emotions
predicted_emotions = []

# Predict emotions
for sentence in tqdm(sentences, desc="Predicting emotions"):
    try:
        if not sentence.strip():
            raise ValueError("Empty sentence")

        # Detect language and get template
        lang = detect_language(sentence)
        hypothesis = hypothesis_templates.get(lang, hypothesis_templates['hi'])

        # Predict emotion using the correct template
        result = emotion_classifier(
            sentence,
            candidate_labels=emotion_labels,
            multi_label=False,
            hypothesis_template=hypothesis
        )

        top_emotion = result["labels"][0]

    except Exception as e:
        print(f"Error with sentence: {sentence[:30]}... | Error: {e}")
        top_emotion = "unknown"

    predicted_emotions.append(top_emotion)

# Add results to dataframe and save
df["Predicted_Emotion"] = predicted_emotions
df.to_csv("emotion_only_results.csv", index=False)
print("Emotion predictions saved to 'emotion_only_results.csv'.")

# Show some results
print(df[["Cleaned_Sentence", "Predicted_Emotion"]].tail(10))



```

### Sentence Embeddings using ai4bharat Model
#### Converting the sentences into a vectors.

```python
## import Libraries
import torch
from tqdm import tqdm
import gc
import os
import shutil
from sklearn.decomposition import IncrementalPCA
from transformers import AutoTokenizer, AutoModel, AutoConfig
import joblib
```

```python



# Load model and tokenizer
model_name = "ai4bharat/indic-bert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()
config = AutoConfig.from_pretrained(model_name)
max_input_length = min(config.max_position_embeddings, 128)
```

```python
#optimized get_embeddings
def get_embeddings(sentences, max_input_length, batch_size=16, device='cuda', n_components=100, chunk_size=2400):
  chunk_dir = 'chunk_embeddings'
  processed_chunks_file = 'processed_chunks.txt'

  #clear previous chunks
  if os.path.exists(chunk_dir):
    shutil.rmtree(chunk_dir)
    print(f"Cleared {chunk_dir}")
  if os.path.exists(processed_chunks_file):
    os.remove(processed_chunks_file)
    print(f"Cleared {processed_chunks_file}")

  os.makedirs(chunk_dir, exist_ok=True)

  total_rows = len(sentences)
  print(f"processing {total_rows} sentences in chunks of {chunk_size}")

  ipca = IncrementalPCA(n_components=n_components, batch_size=chunk_size)
  chunk_files = []

  for chunk_start in tqdm(range(0, total_rows, chunk_size), desc="Processing Chunks"):
      chunk_sentences = sentences[chunk_start:chunk_start + chunk_size]
      chunk_embeddings = []


      for i in tqdm(range(0, len(chunk_sentences), batch_size),desc="Processing Batches", leave=False):
          batch_sentences = chunk_sentences[i:i+ batch_size]
          encoded_input = tokenizer(
              batch_sentences,
              padding=True,
              truncation=True,
              max_length = max_input_length,
              return_tensors ="pt"
      ).to(device)
          with torch.no_grad():
            outputs = model(**encoded_input)
            batch_embeddings = outputs.last_hidden_state[:,0,:].cpu().detach().numpy()
          chunk_embeddings.extend(batch_embeddings)
          del encoded_input, outputs
          torch.cuda.empty_cache()
          gc.collect()


      chunk_embeddings = np.array(chunk_embeddings)
      if chunk_embeddings.shape[0] > 0:
        ipca.partial_fit(chunk_embeddings)

      chunk_filename = os.path.join(chunk_dir, f'indicbert_chunk_{chunk_start}.npy')
      np.save(chunk_filename, chunk_embeddings)
      chunk_files.append(chunk_filename)


      with open(processed_chunks_file,'a') as f:
        f.write(f"{chunk_start}\n")


  print("Applying IncrementalPCA to transform embeddings")
  reduced_embeddings = []
  for chunk_file in tqdm(chunk_files, desc="Trasnformaing Chunks"):
      chunk_data = np.load(chunk_file)
      if chunk_data.shape[0] > 0:
         reduced_chunk = ipca.transform(chunk_data)
         reduced_embeddings.extend(reduced_chunk)


  reduced_embeddings = np.array(reduced_embeddings)
  print(f"Reduced embeddings shape: {reduced_embeddings.shape}")
  print(f"Explained variance ratio: {sum(ipca.explained_variance_ratio_):.4f}")
  joblib.dump(ipca, 'ipca_model.pkl')
  return reduced_embeddings, ipca

sentences = df1['Cleaned_Sentence'].tolist() # Assigning the sentences to the variable

  #Generate embeddings

embeddings_array, ipca =get_embeddings(
         sentences,
         max_input_length,
         batch_size=16,
         device=device,
         n_components=100,
         chunk_size=2400
)

## convert embeddings to Dataframe
embeddings_df = pd.DataFrame(embeddings_array)


```

```python


import torch
import numpy as np
import pandas as pd
from torch.utils.data import Dataset
from torch import nn
from transformers import AutoTokenizer, AutoModel, Trainer, TrainingArguments, DataCollatorWithPadding
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
from sklearn.utils.class_weight import compute_class_weight



# Validate emotions
expected_emotions = {'anger', 'fear', 'joy', 'love','sadness', 'surprise'}
assert expected_emotions == set(balanced_final_df['Predicted_Emotion'].unique()), "Emotions mismatch"

# Encode text and labels
texts = balanced_final_df['Cleaned_Sentence'].values
sentiments = balanced_final_df['Sentiment'].values
emotions = balanced_final_df['Predicted_Emotion'].values

sentiment_encoder = LabelEncoder()
emotion_encoder = LabelEncoder()

sentiment_labels = sentiment_encoder.fit_transform(sentiments)
emotion_labels = emotion_encoder.fit_transform(emotions)

# Compute class weights
sentiment_class_weights = torch.tensor(
    compute_class_weight('balanced', classes=np.unique(sentiment_labels), y=sentiment_labels),
    dtype=torch.float
)
emotion_class_weights = torch.tensor(
    compute_class_weight('balanced', classes=np.unique(emotion_labels), y=emotion_labels),
    dtype=torch.float
)

# Train-test split
X_train, X_test, y_sentiment_train, y_sentiment_test, y_emotion_train, y_emotion_test = train_test_split(
    texts, sentiment_labels, emotion_labels, test_size=0.2, random_state=42, stratify=emotion_labels
)

# Tokenizer
MODEL_NAME = 'ai4bharat/indic-bert'
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# Dataset
class MultiTaskDataset(Dataset):
    def __init__(self, texts, sentiment_labels, emotion_labels, tokenizer, max_length=128):
        self.texts = texts
        self.sentiment_labels = sentiment_labels
        self.emotion_labels = emotion_labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        encoding = self.tokenizer(
            text, padding='max_length', truncation=True, max_length=self.max_length, return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].squeeze(0),
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'sentiment_labels': torch.tensor(self.sentiment_labels[idx], dtype=torch.long),
            'emotion_labels': torch.tensor(self.emotion_labels[idx], dtype=torch.long)
        }

train_dataset = MultiTaskDataset(X_train, y_sentiment_train, y_emotion_train, tokenizer)
eval_dataset = MultiTaskDataset(X_test, y_sentiment_test, y_emotion_test, tokenizer)

# Data collator
#Custom collator that pads sequences and groups sentiment and emotion labels into batches.
class MultiTaskDataCollator:
    def __init__(self, tokenizer):
        self.collator = DataCollatorWithPadding(tokenizer=tokenizer)

    def __call__(self, features):
        sentiment_labels = [f.pop('sentiment_labels') for f in features]
        emotion_labels = [f.pop('emotion_labels') for f in features]
        batch = self.collator(features)
        batch['sentiment_labels'] = torch.stack(sentiment_labels)
        batch['emotion_labels'] = torch.stack(emotion_labels)
        return batch

data_collator = MultiTaskDataCollator(tokenizer)

# Model
class MultiTaskIndicBert(nn.Module):
    def __init__(self, model_name, num_sentiment_labels, num_emotion_labels, sentiment_class_weights=None, emotion_class_weights=None):
        super(MultiTaskIndicBert, self).__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(0.3)
        self.sentiment_classifier = nn.Linear(self.bert.config.hidden_size, num_sentiment_labels)
        self.emotion_classifier = nn.Linear(self.bert.config.hidden_size, num_emotion_labels)
        self.sentiment_loss_fn = nn.CrossEntropyLoss(weight=sentiment_class_weights) if sentiment_class_weights is not None else nn.CrossEntropyLoss()
        self.emotion_loss_fn = nn.CrossEntropyLoss(weight=emotion_class_weights) if emotion_class_weights is not None else nn.CrossEntropyLoss()

    def forward(self, input_ids=None, attention_mask=None, sentiment_labels=None, token_type_ids=None,emotion_labels=None):

       outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask,token_type_ids=token_type_ids)
       pooled_output = self.dropout(outputs.pooler_output)
       sentiment_logits = self.sentiment_classifier(pooled_output)
       emotion_logits = self.emotion_classifier(pooled_output)

       loss = None
       if sentiment_labels is not None and emotion_labels is not None:
          sentiment_loss = self.sentiment_loss_fn(sentiment_logits, sentiment_labels)
          emotion_loss = self.emotion_loss_fn(emotion_logits, emotion_labels)
          loss = sentiment_loss + emotion_loss

       return (loss, (sentiment_logits, emotion_logits))



# Trainer
class MultiTaskTrainer(Trainer):
  def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
    sentiment_labels = inputs.pop("sentiment_labels")
    emotion_labels = inputs.pop("emotion_labels")

    loss, logits = model(**inputs, sentiment_labels=sentiment_labels, emotion_labels=emotion_labels)

    return (loss, {"logits": logits}) if return_outputs else loss


  def prediction_step(self, model, inputs, prediction_loss_only, ignore_keys=None):
    sentiment_labels = inputs.pop("sentiment_labels", None)
    emotion_labels = inputs.pop("emotion_labels", None)

    with torch.no_grad():
        loss, (sentiment_logits, emotion_logits) = model(**inputs, sentiment_labels=sentiment_labels, emotion_labels=emotion_labels)
        loss = loss.mean().detach() if loss is not None else None

    if prediction_loss_only:
        return (loss, None, None)

    return (
        loss,
        (sentiment_logits.detach(), emotion_logits.detach()),  #return torch tensors
        (sentiment_labels.detach(), emotion_labels.detach())   #return torch tensors
    )

# Compute metrics
def compute_metrics(eval_preds):
    preds, labels = eval_preds

    # If preds is a tuple, unpack it
    if isinstance(preds, tuple):
        sentiment_logits, emotion_logits = preds
    else:
        sentiment_logits, emotion_logits = preds[0], preds[1]

    # Convert logits to predictions
    sentiment_preds = np.argmax(sentiment_logits, axis=1)
    emotion_preds = np.argmax(emotion_logits, axis=1)

    # If labels is a tuple, unpack it
    if isinstance(labels, tuple):
        sentiment_labels, emotion_labels = labels
    else:
        sentiment_labels, emotion_labels = labels[0], labels[1]

    # Ensure predictions and labels are NumPy arrays
    sentiment_preds = np.array(sentiment_preds)
    emotion_preds = np.array(emotion_preds)
    sentiment_labels = np.array(sentiment_labels)
    emotion_labels = np.array(emotion_labels)

    # Compute metrics
    sentiment_accuracy = accuracy_score(sentiment_labels, sentiment_preds)
    emotion_accuracy = accuracy_score(emotion_labels, emotion_preds)
    sentiment_f1 = f1_score(sentiment_labels, sentiment_preds, average='weighted')
    emotion_f1 = f1_score(emotion_labels, emotion_preds, average='weighted')
    sentiment_precision = precision_score(sentiment_labels, sentiment_preds, average ='weighted')
    emotion_precision = precision_score(emotion_labels, emotion_preds, average='weighted')
    sentiment_recall = recall_score(sentiment_labels,sentiment_preds, average='weighted')
    emotion_recall = recall_score(emotion_labels,emotion_preds, average='weighted')
    return {
        'sentiment_accuracy': float(sentiment_accuracy),
        'emotion_accuracy': float(emotion_accuracy),
        'sentiment_f1': float(sentiment_f1),
        'emotion_f1': float(emotion_f1),
        'sentiment_precision' : float(sentiment_precision),
        'emotion_precision' : float(emotion_precision),
        'sentiment_recall' : float(sentiment_recall),
        'emotion_recall' : float(emotion_recall)
    }

# Training args
training_args = TrainingArguments(
    output_dir='./multitask_results',
    eval_strategy='epoch',
    save_strategy='epoch',
    num_train_epochs=5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./multitask_logs',
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model='eval_loss',
)

# Model and Trainer init
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultiTaskIndicBert(MODEL_NAME, len(sentiment_encoder.classes_), len(emotion_encoder.classes_),
                           sentiment_class_weights.to(device), emotion_class_weights.to(device)).to(device)

trainer = MultiTaskTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Train
print("Starting training...")
trainer.train()
print("Training complete!")

# Inference function
def predict(text, tokenizer, model, device):
    model.eval()
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        _, (sentiment_logits, emotion_logits) = model(**inputs)
        sentiment_pred = torch.argmax(sentiment_logits, dim=1).item()
        emotion_pred = torch.argmax(emotion_logits, dim=1).item()
    return sentiment_pred, emotion_pred

# wandDB API Key  4c56d758b73803706dfa5c09c1f135337a4710e0
```
